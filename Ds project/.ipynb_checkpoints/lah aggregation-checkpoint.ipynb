```python
import pandas as pd
```

```python
lah = pd.read_excel("pakistan_aqi_weather.xlsx", sheet_name = "Lahore")
```

```python
lah['time'] = pd.to_datetime(lah['time'])
```

```python
print(lah["time"].dtype)
```

```python
hourly = lah.groupby(lah['time'].dt.hour)['pm2_5'].mean()
```

```python
hourly.tail()
```

```python
daily = lah.groupby(lah['time'].dt.date)['pm2_5'].mean()
```

```python
daily.tail()
```

```python
monthly = lah.groupby(lah['time'].dt.to_period('M'))['pm2_5'].sum()
print(monthly.tail())
```

```python
yearly = lah.groupby(lah['time'].dt.year)['pm2_5'].mean()
yearly.head()
```

```python
yearly = lah.groupby(lah['time'].dt.year)['pm2_5'].mean()
yearly.head()
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

lah['season'] = lah['time'].dt.month.apply(get_season)

# Seasonal aggregation
seasonal_summary = lah.groupby('season')[
    [
        'pm2_5',
        'pm10',
        'temperature_2m',
        'relative_humidity_2m',
        'wind_speed_10m',
        'rain'
    ]
].agg(['mean', 'max', 'min'])

print(seasonal_summary)
```

```python
daily_avg = lah.groupby(lah['time'].dt.date)[
    ['pm2_5', 'pm10', 'temperature_2m']
].mean()

daily_avg.head()
```

```python
lah['month'] = lah['time'].dt.month

seasonal = lah.groupby('month')[
    ['pm2_5', 'temperature_2m', 'relative_humidity_2m']
].mean()

seasonal.head()
```

```python
weather_pollution = lah.groupby(
    pd.cut(lah['temperature_2m'], bins=5)
)['pm2_5'].mean()

weather_pollution.head()
```

```python
max_pollution = lah.groupby(
    lah['time'].dt.date
)['pm2_5'].max()
max_pollution.head()
```

```python
rain_effect = lah.groupby(
    lah['rain'] > 0
)['pm2_5'].mean()

rain_effect.head()
```

```python
summary = lah.groupby(lah['time'].dt.month)['pm2_5'].agg([
    'mean',
    'max',
    'min',
    'std'
])

print(summary.head())
```
