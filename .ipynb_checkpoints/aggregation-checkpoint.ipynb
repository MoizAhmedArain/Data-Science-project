```python
import pandas as pd
```

```python
kar = pd.read_excel("pakistan_aqi_weather.xlsx", sheet_name = "Karachi")
```

```python
kar['time'] = pd.to_datetime(kar['time'])
```

```python
print(kar["time"].dtype)
```

```python
#kar = kar.set_index("time")
```

```python
hourly = kar.groupby(kar['time'].dt.hour)['pm2_5'].mean()
```

```python
hourly.tail()
```

```python
daily = kar.groupby(kar['time'].dt.date)['pm2_5'].mean()
```

```python
daily.tail()
```

```python
monthly = kar.groupby(kar['time'].dt.to_period('M'))['pm2_5'].sum()
print(monthly.tail())
```

```python
monthly
```

```python
yearly = kar.groupby(kar['time'].dt.year)['pm2_5'].mean()
yearly.head()
```

```python
yearly = kar.groupby(kar['time'].dt.year)['pm2_5'].mean()
yearly.head()
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

kar['season'] = kar['time'].dt.month.apply(get_season)

# Seasonal aggregation
seasonal_summary = kar.groupby('season')[
    [
        'pm2_5',
        'pm10',
        'temperature_2m',
        'relative_humidity_2m',
        'wind_speed_10m',
        'rain'
    ]
].agg(['mean', 'max', 'min'])

print(seasonal_summary)
```

```python
daily_avg = kar.groupby(kar['time'].dt.date)[
    ['pm2_5', 'pm10', 'temperature_2m']
].mean()

daily_avg.head()
```

```python
monthly_avg = kar.groupby(kar['time'].dt.to_period('M'))[
    ['pm2_5', 'pm10', 'ozone']
].mean()

monthly_avg.head()
```

```python
kar['month'] = kar['time'].dt.month

seasonal = kar.groupby('month')[
    ['pm2_5', 'temperature_2m', 'relative_humidity_2m']
].mean()

seasonal.head()
```

```python
weather_pollution = kar.groupby(
    pd.cut(kar['temperature_2m'], bins=5)
)['pm2_5'].mean()

weather_pollution.head()
```

```python
wind_pollution = kar.groupby(
    pd.cut(kar['wind_speed_10m'], bins=5)
)['pm2_5'].mean()

wind_pollution.head()
```

```python
max_pollution = kar.groupby(
    kar['time'].dt.date
)['pm2_5'].max()
max_pollution.head()
```

```python
rain_effect = kar.groupby(
    kar['rain'] > 0
)['pm2_5'].mean()

rain_effect.head()
```

```python
summary = kar.groupby(kar['time'].dt.month)['pm2_5'].agg([
    'mean',
    'max',
    'min',
    'std'
])

print(summary.head())
```
