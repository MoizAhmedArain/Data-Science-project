```python
p = d = q = [0, 1, 2]
P = D = Q = [0, 1,2]
s = [12] # Keep seasonal cycle fixed at 12 months

# Generate all unique combinations of non-seasonal and seasonal parameters
pdq_combinations = list(itertools.product(p, d, q))
seasonal_combinations = list(itertools.product(P, D, Q, s))

best_accuracy = -float('inf')
best_pdq = None
best_seasonal = None
best_forecast = None

print("Searching for the highest accuracy combination...")

# 2. Run the Optimization Loop
for pdq in pdq_combinations:
    for seasonal in seasonal_combinations:
        try:
            # Fit the model temporary with current parameter settings
            model = SARIMAX(y_train,
                            exog=X_train,
                            order=pdq,
                            seasonal_order=seasonal,
                            enforce_stationarity=False,
                            enforce_invertibility=False)
            results = model.fit(disp=False)
            
            # Predict over test window
            pred = results.forecast(steps=len(test), exog=X_test)
            
            # Calculate accuracy percentage (100 - MAPE)
            mape = np.mean(np.abs((y_test.values - pred.values) / y_test.values)) * 100
            accuracy = 100 - mape
            
            # If current combination is higher than our previous best, save it!
            if accuracy > best_accuracy:
                best_accuracy = accuracy
                best_pdq = pdq
                best_seasonal = seasonal
                best_forecast = pred
                
        except:
            continue

print("\n=========================================")
print("         OPTIMIZATION COMPLETED          ")
print("=========================================")
print(f" Highest Accuracy Found : {best_accuracy:.2f}%")
print(f" Best non-seasonal order: {best_pdq}")
print(f" Best seasonal order    : {best_seasonal}")
print("=========================================")

# 3. Restructure and visualize the highest accuracy results immediately
test_df = pd.DataFrame({
    'time': test.index,
    'pm2_5': y_test.values.flatten(),
    'Type': 'Actual Test Data'
})

forecast_df = pd.DataFrame({
    'time': test.index,
    'pm2_5': best_forecast.values.flatten(),
    'Type': f'Optimized Prediction ({best_accuracy:.1f}% Accurate)'
})

plot_df = pd.concat([test_df, forecast_df], ignore_index=True)

# 4. Generate the visually optimized Altair Chart
alt.data_transformers.disable_max_rows()

chart = alt.Chart(plot_df).mark_line(point=True).encode(
    x=alt.X('time:T', title='Date'),
    y=alt.Y('pm2_5:Q', title='PM2.5 Level', scale=alt.Scale(zero=False)),
    color=alt.Color('Type:N', title='Legend', scale=alt.Scale(
        domain=['Actual Test Data', f'Optimized Prediction ({best_accuracy:.1f}% Accurate)'],
        range=['#2ca02c', '#d62728']
    )),
    strokeDash=alt.StrokeDash('Type:N', scale=alt.Scale(
        domain=['Actual Test Data', f'Optimized Prediction ({best_accuracy:.1f}% Accurate)'],
        range=[[0], [4, 4]]
    )),
    tooltip=[alt.Tooltip('time:T', title='Date', format='%B %Y'), alt.Tooltip('pm2_5:Q', title='PM2.5', format='.2f')]
).properties(
    title=f'Highest Matching Performance Graph (Accuracy: {best_accuracy:.2f}%)',
    width=800,
    height=400
).interactive()

chart
```
